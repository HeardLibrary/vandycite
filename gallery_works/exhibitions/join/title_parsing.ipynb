```python
import pandas as pd
exhibited_works = pd.read_csv('artwork_in_exhibitions.csv', dtype=str)
exhibited_works
```

```python
exhibitions_list = list(exhibited_works['exhibitions'])
```

```python
split_exhibitions = []
for exhibition in exhibitions_list:
    sub_list = exhibition.split('|')
    for item in sub_list:
        split_exhibitions.append(item.strip())
split_exhibitions
```

```python
unique_exhibitions_list = list(set(split_exhibitions))
len(unique_exhibitions_list)
```

```python
exhibitions_df = pd.DataFrame({'name': unique_exhibitions_list})
exhibitions_df
```

```python
exhibitions_df = exhibitions_df.sort_values(by=['name'])
exhibitions_df.to_csv('exhibitions.csv', index = False)
```

```python
exhibitions_df = pd.read_csv('exhibitions_raw.csv', dtype=str)
exhibitions_df
```

```python
split_names = exhibitions_df['gallery_raw_string'].str.split(pat='(', n = 1, expand = True)
exhibitions_df['title'] = split_names[0]
split_dates = exhibitions_df['raw_string'].str.split(pat='(', n = 1, expand = True)
dates_with_paren = split_dates[1]
exhibitions_df['dates'] = dates_with_paren.str.split(pat=')', n = 1, expand = True)[0]
exhibitions_df
```

```python
start_end = exhibitions_df['dates'].str.split(pat='-', n = 1, expand = True)
exhibitions_df['start'] = start_end[0].str.strip()
exhibitions_df['end'] = start_end[1].str.strip()
exhibitions_df
```

```python
exhibitions_df.to_csv('exhibitions_parsed.csv', index = False)
```

```python
from datetime import datetime
exhibitions_df = pd.read_csv('exhibitions_parsed.csv', dtype=str)
exhibitions_df = exhibitions_df.fillna('')
exhibitions_df
```

```python
output_list = []
for index, work in exhibitions_df.iterrows():
    output_dict = {}
    if work['title'] == '':
        continue
    output_dict['title'] = str(work['title']).strip()
    output_dict['title'] = output_dict['title'].replace('  ', ' ')
    if ',' in work['start']:
        date = str(datetime.strptime(work['start'], '%B %d, %Y'))[:10]
    elif ' ' in work['start']:
        date = str(datetime.strptime(work['start'], '%B %Y'))[:7]
    else:
        date = work['start']
    output_dict['start'] = date
    if ',' in work['end']:
        date = str(datetime.strptime(work['end'], '%B %d, %Y'))[:10]
    elif ' ' in work['end']:
        date = str(datetime.strptime(work['end'], '%B %Y'))[:7]
    else:
        date = work['end']
    output_dict['end'] = date
    output_dict['raw_string'] = work['raw_string']
    output_list.append(output_dict)
output_df = pd.DataFrame(output_list)
output_df.to_csv('exhibitions_clean.csv', index = False)
    
```
